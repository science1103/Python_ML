# 에러, 손실, 비용함수의 모양이 다음과 같다고 가정할 때

$$
    f(x) = x^2 - 4x+ 6
$$

- 1변수일 때, 그림을 그리기 제일 쉽고, 
- 손실함수에서 함수의 값이 가장 작게하는 x를 찾고 싶은 것 
- 예시.1 

```python
#일변수 일때
def f(x):
    return x**2 - 4*x + 6 
```

```python
x = np.linspace(-5, 5, 101)
y = f(x)
```

```python
# 이런 모양의 손실함수가 등장 
plt.plot(x,y)
plt.show()
```

```python
def grad_fx(x):
    return 2*x - 4
```

### 러닝 레이트를 최소화한 값

```python
x0 = 0 # x의 초기값을 0으로 지정 
learning_rate = 0.01
maxIter = 200
paths = []

for it in range(maxIter):
    paths.append(x0)
    x1 = x0 - learning_rate * grad_fx(x0)
    #print("횟수: {}, argx: {}, slope: {}, F value: {}".format(it, x1, grad_fx(x1), f(x1)))
    x0 = x1 
```

```python
x = np.linspace(0,2.5,10000)
paths = np.array(paths) #x의 기울기가 learning_Rate만큼 이동하는 변화하는 x값을 담아내어 그래프에 표현하려는 것 
plt.grid()
plt.plot(x,f(x))
plt.plot(paths, f(paths), 'o-')
plt.show()
```

손실함수의 함수값(x)이 가장 작아진다는 것은 오차가 작아진다는 뜻 = 올바르게 찾아가고 있다

```python
#           x값
plt.plot(f(paths), '-o')
plt.xlabel('x')
plt.ylabel('loss')
plt.show()
# 줄어들고 있는 손실 값 = 손실을 최소화하고 있다 
```

## non-convex function 

$$
    \min_x x sin(x)
$$

- convex하지 않은 함수를 하나 가정하겠다
- 손실(에러)함수가 이런 모양이라면 (가정)
- 일변수인데도 불구하고 이렇게 나오며
- 이런 그래프는 여러가지 문제가 발생함

```python
x = np.linspace(-3, 10, 501)
f = lambda x:x * np.sin(x)
y = f(x)
```

```python
plt.plot(x,y)
plt.show()
```

그라디언트 디센트 값은 기울기가 0인 값을 최저값이라 판단하지만 위 그래프의 기울기가 0이되는 곳은 최소 5개이며, \
오류값이 글로벌 멕시멈인 8 또한 기울기가 0이기 때문에 문제가 있다.

```python
grad_fx = lambda x: np.sin(x) + x * np.cos(x)
```

```python
x0 = 7.0
learning_rate = 0.1
maxIter = 10
paths = []

for it in range(maxIter):
    paths.append(x0)
    x1 = x0 - learning_rate * grad_fx(x0)
    #print("횟수: {}, argx: {}, slope: {}, F value: {}".format(it, x1, grad_fx(x1), f(x1)))
    x0 = x1 
```

```python
# 찾고자하는 글로벌 미니멈 파악하는 방법 
paths = np.array(paths)
plt.plot(x, y)
plt.grid()
plt.plot(paths, f(paths), 'o-')
plt.show()
```

```python
plt.plot(f(paths), 'o-')
plt.show()
```

## 변수가 2개인 경우 
- 그림을 그리는 것에만 집중할 수 있도록

- 그림을 그려볼 수 있는 마지막 
- visualize.py를 가져다가 쓸 것임 

$$
    f(x,y) = (x-2)^2 + (y-2)^2
$$

```python
from visualize import *
```

```python
def f(x, y):   # 함수 f 값 
    return (x-2) ** 2 + (y-2) ** 2 

def grad_fx(x, y): # 그에 맞는 편미분 x 
    return 2 * x - 4 
    #return 2*(x-2)
    
def grad_fy(x, y): # 그에 맞는 편미분 y 
    return 2 * y - 4 
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 25), np.linspace(-5, 5, 25))
z = f(x, y)
```

```python
argmin = np.array([2,2])
argmin = argmin.reshape(2, 1)
argmin
```

```python
surf(f, x, y, minima=argmin)
```

```python
# 3차원을 2차원으로 옮긴 곡선 #그라디언트를 인자로 넘겨줬기 때문에 그래프에서 하강하는 곡선의 방향성이 나타나는 것 
contour_with_quiver(f,x,y,grad_fx,grad_fy,minima=argmin)
```

```python
x0 = np.array([4,-4]) #열벡터
learning_rate = 0.2
maxIter = 10
paths = [x0]

for it in range(maxIter):
    paths.append(x0)
    grad = np.array([grad_fx(*x0), grad_fy(*x0)])
    x1 = x0 - learning_rate * grad
    x0 = x1 
    
paths = np.array(paths)
paths = np.array(np.matrix(paths).T)
```

```python
contour_with_path(f, x, y, paths, minima=np.array([[2],[2]]))
```
