```python
import warnings 
warnings.filterwarnings('ignore')
warnings.filterwarnings('always')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.colors import ListedColormap
import seaborn as sns
```

이 곳에서도 마찬가지로 bike_train.csv 파일을 불러와 데이터를 조작한다. 

```python
rd = pd.read_csv('./bike_train.csv')
```

```python
# 타입 바꾸기 => datetime.type으로 
rd['datetime'] = pd.to_datetime(rd['datetime'])
```

```python
rd['date_time'] = rd['datetime']
```

```python
rd.set_index('date_time', inplace=True)
```

```python
rd['datetime']
```

```python
            # datetime은 dt라는 속성값을 지니는 데 이를 통해 연월일을 나눌 수 있다.
rd['year']=rd.datetime.dt.year
```

```python
rd['month']=rd.datetime.dt.month 
```

```python
rd['day']=rd.datetime.dt.day
```

```python
rd['hour']=rd.datetime.dt.hour
```

```python
rd['dow']=rd.datetime.dt.dayofweek
```

```python
data = rd['count']
sns.distplot(data, kde=True)            # datetime은 dt라는 속성값을 지니는 데 이를 통해 연월일을 나눌 수 있다.
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='season', data=rd)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='holiday', data=rd)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='weather', data=rd)
plt.show()
```

```python
plt.figure(figsize=(20,6))
                                       # 색 속성  
sns.countplot(x='temp_group', data=rd, palette='Set3')
plt.show()
```

다시 자전거 데이터로 돌아가기

## 다차원 데이터
- 2차원의 연속형(실수값) 변수의 경우

```python
plt.figure(figsize=(12,6))
sns.jointplot(data=rd, y='count', x='temp', kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pairplot(data=rd)
plt.show()
```

## 2차원 복합 데이터
- 범주형과 실수형 

```python
plt.figure(figsize=(12,3))
sns.boxplot(data=rd, y='count', x='weather', palette='rocket')
```

```python
# 바이올린 플롯 실제 데이터의 형상 
plt.figure(figsize=(12,5))
sns.violinplot(data=rd, y='count', x='weather', palette='rocket')
```

```python
# 바이올린 플롯 실제 데이터의 형상 
plt.figure(figsize=(12,10))
sns.stripplot(data=rd, y='count', x='weather', hue='workingday', palette='rocket')
```

hue = 색으로 나타내는 범주 

```python
plt.figure(figsize=(20,9))
sns.swarmplot(data=rd, y='count', x='hour', hue='workingday', palette='rocket')
plt.show()
```
