## 다항함수로 접근해보기 

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
x = np.array([20,21,22,23,24])
y = np.array([15,16,21,33,42])
```

```python
plt.plot(x,y, 'o')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## loss function 

- MSE (MEAN Squared Error)

$$
        loss(w) = \frac{1}{N} \sum_{i=1}^N | w_0 + w_1x^1 + w_2x^2 + w_3x^3 - y_i | ^ 2 
$$

- $ w_0x_i^2 + w_1x_i + w_2 $ : 일반적인 함수
- $w$ :  모수(값을 찾아야 함)

```python
def loss(w):
    return np.mean((w[0] + w[1]*x + w[2]*x**2 + w[3]*x**3 - y)**2)
```

```python
w0 = [0,0,0,0]
ret = minimize(loss, w0)
```

```python
w = ret.x 
y_predict = w[0] + w[1] * x + w[2] * x **2 + w[3] * x ** 3 
print(w)
```

```python
ret.x
```

```python
plt.plot(x,y, 'o')
plt.plot(x, y_predict, alpha = 0.5, lw = 2 )
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
tmp1 = np.polyfit(x, y, 1)
tmp4 = np.polyfit(x, y, 4)

lin = np.poly1d(tmp1)
lin4 = np.poly1d(tmp4)
```

```python
plt.plot(x,y, 'o')
plt.plot(x, lin(x))
#plt.plot(x, lin2(x), '--')
#plt.plot(x, lin3(x), '--' )
plt.plot(x, lin4(x), alpha=1, lw=2)
plt.grid()
plt.show()
```
