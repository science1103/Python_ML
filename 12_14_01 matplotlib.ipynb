```python
# 쓸데없는 오류 원인 보기 싫을 때만 사용
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore') 

import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.font_manager
```

## 한 화면에 여러 그래프 출력하기 
- subplot  

```python
plt.figure(figsize=(20,10))
#      nrows, ncols, index
plt.subplot(2, 2, 1)
plt.plot(x,y,'r')

           # 2행 2열 크기로 결과값 2번쨰에 위치
plt.subplot(2,2,2)
plt.plot(y, x, 'g')

plt.subplot(2,2,3)
plt.plot(x, x**2, 'b' )

plt.subplot(2, 2, 4)
plt.plot(x, np.log(x))
```

# matplotlib 의 또 다른 그림들

## 막대그래프
- 바차트, 히스토그램, ... 

```python
x = np.arange(5)
xlabel = ['첫째', '둘째', '셋째', '넷째', '다섯째' ]
y = [100, 200, 300, 400, 500]
                # x축의 기준선 center가 디폴트      축의 테두리 색 테두리 두께               
plt.bar(x, y, width=0.7, align= 'edge', color='r', edgecolor='b', linewidth=3)
   # 값을 할당해주는 함수, 레이블 할당도 가능함 
plt.xticks(x, xlabel)
plt.show()
```

```python
x = np.arange(5)
y = [100, 200, 300, 400, 500]
#              =width                                                             # 투명도
plt.barh(x, y, height=0.5, align= 'edge', color='r', edgecolor='b', linewidth=3, alpha=0.5)
plt.xlabel('일단x')
plt.show()
```

## 히스토그램 그리기
- y가 따로 필요없기 때문에, 단변량 그림그리기에 상당히 좋음
- bins는 히스토그램의 bar의 개수를 설정하는 것이다. 

```python
data = np.random.randint(1, 100, size=200)
len(data)
```

```python
#                   나눌 수 
plt.hist(data, bins=201)
plt.show()
```

## Scatter(산점도)

```python
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
area = (20 * np.random.rand(N))**2
color = np.sqrt(area)
plt.scatter(x,y, s=area, c=color, alpha = 0.8)
plt.show()
```
