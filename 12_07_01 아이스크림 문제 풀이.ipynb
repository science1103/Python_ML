## 선형적인 관계로 가정(모수적 추정)

- $y = ax + b$
- 주어진 데이터를 가장 잘 설명할 수 있는 선을 찾는 것

## error function (오차함수) or Cost Function

- Squared Error 
- Mean Squared Error (MSE)

$$ 
    \sum_{i = 1}^n | y_1 - (ax_i + b) | ^2 
$$

$$
    \min_{a,b}^n | y_1 - (ax_i + b) | ^2     
$$

1. 주어진 선 : $ax + b$ 
2. 기준: squared Error 
3. 좋은 상황: at least 
4. 선을 찾는 것 : $a, b$
5. 오차를 최소로 하는 a,b값을 찾고 싶다.
$$
    (a,b) = argmin_{a,b}
$$

## 2차 함수의 최솟값을 구하기 

$$
    f(x) = x^2 - 4x + 6 
$$

```python
def f(x):
    return x**2 - 4*x + 6 
```

```python
x = np.linspace(-5, 5, 101)
print(x)
```

```python
y = f(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Y값을 최소로 하는 x가 존재하는가에 대한 함수
xidx = np.argmin(y)
argX = x[xidx]
valueF = f(argX)
print('argX : ', argX) #2에 가까운 숫자가 최소값
print('f min value : ', valueF)
```

```python
plt.plot(x,y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.plot(argX, valueF, 'xy')
plt.show()
```

## 최급하강법(Steepest Descent) or 경사하강법(Gradient Descent)

$$
    x_{k+1} = x_k - \alpha \nabla f(x_k) \\
    \\
    \\
    x^{k+1} = x^{k} - \alpha \nabla f(x^{k})
$$

- $\nabla$ = gradient 
- $a$ = learning rate
- $x^k$ = x의 k번째
- $\nabla f(x_k)$ = x의 k번째 기울기
- 기울기가 최소로 낮아지는 곳인 x가 근사하는 곳을 찾아간다. 
- 접선의 기울기($a$) 작아진다 = 함수의 값이 작아진다. 

```python
def grad_fx(x):
    return 2*x + 4
```

```python
x0 = 0
learning_rate = 0.01
maxIter = 10000

for it in range(maxIter):
    x1 = x0 - learning_rate * grad_fx(x0)
    print("횟수: {}, argx: {}, slope: {}, F value: {}".format(it, x1, grad_fx(x1), f(x1)))
    x0 = x1 
```
