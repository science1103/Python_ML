## 미분이란(Differemtials)?
- 머신러닝/딥러닝 알고리즘에서 사용하는 최적화 기법(알고리즘)의 핵심
- 함수의 기울기 : 함수의 변화량을 나타냄 

```python
def f(x): 
    return 2 * x

# f = lambda x; 2 * + 6
```

```python
#선형함수
x = np.linspace(0, 5, 100)
print(x)
```

```python
y = f(x) 
print(y)
```

```python
plt.plot(x,y)
plt.grid()
plt.xlim(0, 4)
plt.ylim(0, 8)
plt.xticks(np.arange(0, 5))
plt.yticks(np.arange(0, 9))

plt.annotate('', xy=(2,2), xytext=(1,2), arrowprops=dict(facecolor='red'))
plt.annotate('', xy=(2,4), xytext=(2,2), arrowprops=dict(facecolor='red'))

plt.annotate('', xy=(3,4), xytext=(2,4), arrowprops=dict(facecolor='red'))
plt.annotate('', xy=(3,6), xytext=(3,4), arrowprops=dict(facecolor='red'))

plt.show()
```

어느 위치에 있던 동일한 기울기를 구할 수 있으며
직선의 경우 기울기의 변화는 거의 없다.

### 곡선의 미분

```python
f = lambda x: x**2 - 1 * x + 2
```

```python
x = np.linspace(0, 5, 100)
y = f(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.annotate('', xy=(2,4), xytext=(1, 2), arrowprops=dict(facecolor='red'))
plt.annotate('', xy=(5,6), xytext=(5, 6), arrowprops=dict(facecolor='red'))
plt.xlabel('x: speed')
plt.ylabel('y: distance')
plt.show()
```

- 직선이 아닌 함수의 변화량(기울기)이 일정치 않기 때문에 변화량을 측정하는 위치마다 기울기 값이 다르게 나올 수 밖에 없다.
- 따라서 우리에게는 평균변화율 값이 필요하다! 

```python
plt.plot(x,y)
plt.grid()
plt.plot(3, 8, 'ro')
plt.plot(4, 14, 'bo')

plt.xlabel('x: speed')
plt.ylabel('y: distance')
plt.show()
```

- 빨간점을 a, 파란점을 b라고 가정하면, ... 
- 구간에 대한 평균변화율은 다음과 같이 정의된다. 

$$
    \frac{\Delta y}{\Delta x } = \frac{f(b) - f(a)}{b - a} = \frac{f(a + \Delta x)-f(a)}{\Delta x}
$$

## 순간 변화율이란? 

- 충분히 작은 구간에서의 평균 변화율 

$$
    \frac{\Delta y}{\Delta x } = \frac{dy}{dx} = \frac{df(x)}{dx} = \frac{d}{dx}f(x)
$$

- $ \Delta x $를 0.01로 놓고 계산한다면?

$$
    df(x) = \frac{f(x+0.01) - f(x)}{0.01}
$$

#### 순간변화율의 예문
- 다음과 같은 함수를 가정해보자 : $ f(x) = x^2 $
- 이 함수의 순간변화율은 $ 2x $이다.
- $x = 3$일때, 순간변화율은?
$$
   \frac{d}{dx}f(3) = 2(3) = 6
$$

$$
    df(3) = \frac{f(3 + 0.01) - f(3)}{0.01} = \frac{3.01^2 - 3^2}{0.01} = 6.01
$$

- 실제 순간변화율과는 오차가 발생한다.
- 즉, 0.01은 순간변화율이라고 볼 수 없다. 
- 0.001, .00001, 0.00000001 ... 0.00000000000000000001 이상으로 변화량을 아주 작게 줄인다해도 오차는 지속적으로 발생한다.
- 그럼 얼마나 작아야 순간변화율이 나타나는 것일까



　
　
 　
  　　
    　
     　     
## 미분의 정의 

- 변화율을 최소화해도 순간변화율을 구할 수 없음을 인지했다.
- 극한(limit)의 개념을 사용(수렴, 극한, 급수, 유리수의 조밀성, 완비성 공리, ... ) - 수업에선 안다룸

$$
     \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
$$


- 분모를 순간에 이를 정도로(극한) 작게 만들면 순간 변화율이라고 지칭하며,
- 극한이 존재한다면, 극한값에 대한 평균 변화율 을 `순간 변화율` 또는 `미분계수` 라고 정의한다. 
- 순간 변화율은 그 위치에서의 접선의 기울기로 해석할 수 있다.
- 미분계수를 값으로 하는 함수를 도함수라고 하며 $f'()$ `f프라임`이라고 표기한다.



- 위의 좌표값을 이용하여 극한의 개념을 알아보자

$$
    f'(a) = \lim_ {b \to a} \frac{f(b)-f(a)}{b-a} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} 
$$

## 다항함수의 미분 

$$
    f(x) = x^2
$$

1. 미분의 정의에 의하여...
$$
    \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} = \lim_{\Delta x \to 0} \frac{(x + \Delta x )^2 - x^2}{\Delta x}\\
    \\
    = \lim_{\Delta x \to 0} \frac{x^2 + 2\Delta x \cdot x + \Delta x^2 - x^2}{\Delta x}= \lim_{\Delta x \to 0} \frac{\Delta x(2x + \Delta x)}{\Delta x} = 2x
$$
2. 델타값이 0으로 수렴하기 때문에 $2x$가 나옴

- 거듭제곱에 대한 미분의 정의

$$
    \frac{d}{dx}{x^n} = nx^{n-1} \\
    \frac{d}{dx}{x^2} = 2x^{2-1} = 2x
$$

## 단일변수/일변수 함수의 미분에 대해 알아보자 

- 미분계수를 sympy 함수를 이용해 알아보는 방법
$$
    f(x) = x^2 + 4x + 6
$$

```python
def h(x):
    return x**2 + 4*x + 6 

def dh(x): 
    return 2 * x + 4
```

```python
x = sp.Symbol('x')
f = x ** 2 + 4 * x + 6 
sp.diff(f, x)
```

### 다항함수와 도함수의 그래프

```python
# 다항함수
a = np.linspace(-5, 5, 100)
fa = h(a)

# 도함수 
dfa = dh(a)
```

```python
plt.plot(a, fa, '-b') # 다항함수의 그래프 
plt.plot(a, dfa, 'r:') # 다항함수의 도함수 그래프
plt.show()
```

### 예제.1 주어진 다항함수의 미분 계수를 구하고 그림을 그려라.


$$
    f(x) = x^3  - 5x 
$$

```python
x = sp.Symbol('x')
b = x ** 3 - 5 * x
bp = sp.diff(b, x)
bp
```

```python
x = sp.Symbol('x')
# 다항함수
def t(x):
    return x**3 - 5*x

#도함수 
def tp(x):
    return 3*x**2 - 5

tp(x)
```

```python
b = np.linspace(-5, 5, 100)
tb = t(b) # 다항함수
tpb = tp(b) # 도함수
```

```python
#.subplot(가로,세로,위치)
plt.subplot(2,1,1)
plt.plot(b, tb)
plt.show()

# 순간의 변화율 = 도함수 
plt.subplot(2,1,1)
plt.plot(b, tpb, 'r:')
plt.show()
```

## Gradient 
- 벡터장
- 편미분값의 벡터 표현

### 2차원 다항함수에 대한 편미분값의 벡터 표현 

$$
    \nablaf = \begin{bmatrix}
    
                \frac{\Delta f}{\Delta x} \\
        
                \frac{\Delta f}{\Delta y}
    \end{bmatrix}
$$

```python
F= sp.Matrix(1,1, [f])
F
```

```python
# x에 따른 y함수의 변화량 
grad = F.jacobian([x,y]).T
```

```python
grad
```
